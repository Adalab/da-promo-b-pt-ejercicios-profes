```python
def llamar_api(url):

    llamada = requests.get(url)

    print(f"el estado de la llamada es: {llamada.status_code}")

    if llamada.status_code != 200:
        print(f"esto ha fallado por {llamada.reason}")
    
    else:
        return llamada.json()
```

```python
llamar_api(url_pokemon)
```
