# Repaso Selenium

## Ejercicio: Controlando el sitio web, exportar una tabla que muestra los nombres y precios de los libros de Stephen King en boocenter.es

>pagina: https://www.bookcenter.es/

```python
#importar librerias

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from time import sleep
```

```python
url = "https://www.bookcenter.es/"

driver = webdriver.Chrome()
driver.maximize_window()
driver.get(url)

driver.find_element("css selector","#cookie-accept").click()
sleep(2)

driver.find_element("css selector","#busqueda > input.search-text").send_keys("stephen king",Keys.ENTER)

diccionario_stephen = {"nombre":[],"precio":[]}

for patata in range(1,4):
    driver.find_element("css selector",f"#main-container > div > div > div.row > ul > li:nth-child({patata})").click()
    print("el ressultado clicked")
    nombre = driver.find_element("css selector","#titulo").text
    precio = driver.find_element("xpath",'//*[@id="main-container"]/div/div[1]/div/div[2]/div[1]/div[1]/div[2]/span').text
    print("hemos sacado el precio")
    diccionario_stephen["nombre"].append(nombre)
    diccionario_stephen["precio"].append(precio)
    driver.back()
    
    
driver.close()
tabla = pd.DataFrame(diccionario_stephen)
tabla

#driver.close()
```

# Repaso BeautifulSoup:
## Ejercicio: utulizando webscrapping, exportar una tabla con nombres de los libros que salen en la primera pagina de bookcenter.es

```python
# Importar librerias BeautifulSoup
from bs4 import BeautifulSoup
```

```python

url = 'https://www.bookcenter.es/'


response = requests.get(url)


html_content = response.content
    

soup = BeautifulSoup(html_content, 'html.parser')
    
soup.prettify()

book_titles = soup.find_all("dd",class_ = 'title')
book_titles
name = []
for i in book_titles:
    name.append(i.find("a").text)
name

df = pd.DataFrame(name)

df
```
