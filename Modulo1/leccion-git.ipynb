## Git & GitHub
Crea. Rastrea. Colabora.


### Git y GitHub de Manera Sencilla
Imagina que estás escribiendo un libro y quieres probar algunas ideas sin arruinar el original. En lugar de hacer cambios directamente en tu único libro, haces copias para experimentar.
Ahora, piensa en Git como un asistente que te ayuda con esas copias y cambios:
Repositorio: Tu libro es como un "repositorio". Es el lugar principal donde guardas todas las versiones.

- Commits: Cada copia o versión de tu libro se llama "commit". Es como tomar una foto de tu libro en ese momento.

- Push y Pull: Cuando estás listo para compartir tus cambios con otros, haces un "push" para enviar tus versiones al lugar central. Si alguien más hizo cambios, haces un "pull" para obtener sus actualizaciones.

- GitHub: Imagina GitHub como una biblioteca en línea donde puedes guardar tu libro. Otros pueden verlo, colaborar contigo y sugerir cambios.


## Git: Control de Versiones Eficiente
Sistema de control de versiones Open Source creado por el creador de Linux, que se instala localmente en el ordenador:
- Registro de Cambios.
- Reversión y Auditoría.


## GitHub: Colaboración y Almacenamiento  en la Nube

Servicio de alojamiento en la nube para proyectos versionados con Git. Proporciona una interfaz de usuario para ver el cambio en los ficheros:
- Almacenamiento en la Nube.
- Colaboración Remota.
- Gestión de Proyectos.
- Contribuciones Abiertas
## Crea un Repositorio en GitHub
1. https://github.com/
2. En tu perfil, haz click en “New” (lado izquierdo-superior).
3. Añade nombre y descripción (si se añade Readme, esta descripción forma parte del Readme por defecto).
4. Elige entre “Public/Private”. En este caso, Public.
5. Haz click en “Add Readme”.
6. Haz click en “Create Repository”.

## Clona tu Repo Localmente: 

1. En Code > HTTPS: copia la Url.
2. En tu Terminal, dentro de la carpeta donde quieras clonar el repo: 
`git clone “URL”` (introduce nombre de usuario y password cuando te lo pida. Si no funciona, es posible que tengas que generar un token).
3. Sitúate dentro del repo e introduce ls. Ahí verás que ficheros existen dentro del repo; en este caso, deberíais ver el fichero README.md.
4. Si haces `git status`, verás que no hay ningún nuevo cambio; todo está al día.

## Invita a una Colaboradora 
(tu compañera de Pair Programming - PP) & Acepta la Invitación Recibida

**Para invitar a un colaborador:**
- En GitHub (dentro del repo) > Settings > Collaborators > Add People -> buscar usuario.

**Para aceptar la invitación:**
- Tendrás un mensaje en tu cuenta de email.



## Resumen de comandos

- Para clonar un repositorio:
  
  ```bash
  git clone url-del-proyecto
  ```

- Para subir los cambios a Github

    ```bash
    git add -A // añadir cambios
    git commit -m "mensaje" // poner mensaje
    git push //subir cambios 
    ```
    
- Para bajar cambios 

    ```bash
    git pull
    ```

## TIPS
- Antes de empezar a trabajar en alguna tarea nueva: asegúrate de hacer git pull para tener los últimos cambios del repo!
- No olvides “guardar” todos tus cambios una vez acabes la tarea o la sesión y subir esos cambios al repo para que tus compañeras trabajen sobre un repo actualizado. Esto significa hacer `add + commit + push` (mirar sintaxis exacta).
- De momento, para evitar problemas, comunicáos entre vosotras cuando hagáis algún cambio y así todas estáis al día de los cambios en el repo.
